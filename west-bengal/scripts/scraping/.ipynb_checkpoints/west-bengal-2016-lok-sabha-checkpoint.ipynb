```python
import requests
from bs4 import BeautifulSoup

import json
```

```python
payloadURL = 'http://myneta.info/westbengal2016/'
```

```python
rawResponse = requests.get(payloadURL)
soupParser = BeautifulSoup(rawResponse.content, 'html.parser')
allTables = soupParser.find_all('table')
```

```python
tableCount= 0
for dataTable in allTables:
    tableHeaders = [header.text for header in dataTable.find_all('th')]
    tableRows = []
    for singleRow in dataTable.find_all('tr'):
        rawRow = list(data.text.replace(' ', '').rstrip() for data in singleRow.find_all('td'))
        if rawRow != []:
            tableRows.append(rawRow)
    
    for foo in tableRows:
        print(foo)
    print('--')
    
    # parse the data to JSON
    myJson = {}
    
    for headingIndex in range(len(tableHeaders)):
        myJson[tableHeaders[headingIndex]] = list(tableRows[headingIndex])
    
    realJson = json.dumps(myJson)
    #print(realJson)
    
    with open(str(tableCount) + '.json', 'w+') as f:
        json.dump(myJson, f)
    
    tableCount += 1
```
