```python
import pandas as pd
from pandas.io.json import json_normalize
import ast
import json
```

```python
df = pd.read_json('./data/wb_2016_winner_expenses_2.json')
df.drop(df.columns[0], axis=1)
#foo = json_normalize(df, 'Candidate')
df = df.T
df = df.iloc[1:]
df.head()
```

```python
v = json_normalize([ast.literal_eval(json.dumps(j)) for j in df['Candidate'].dropna()], sep='_')
v.index = df.index[df.pop('Candidate').notna()]
df = df.join(v)
```
